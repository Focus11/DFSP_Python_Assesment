```python
# A strings of code that can carry out mathemeatical operations on integers variables x and y
x = 15
y = 5
addition = x + y

print(f'Sum of variables: {addition}')

subtraction = x - y

print(f'Difference: {subtraction}')

multiplication = x * y

print(f'Products: {multiplication}')

division = x / y

print(f'Division: {division}')
```

```python
# A strings of code that can carry out mathemeatical operations on float a and b
a = 12.5
b = 3

addition = a + b

print(f'Sum of variables: {addition}')

subtraction = a - b

print(f'Difference: {subtraction}')

multiplication = a * b

print(f'Products: {multiplication}')

division = a / b

print(f'Division: {division}')
```
